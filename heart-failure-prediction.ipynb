# **Importing Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.naive_bayes import GaussianNB 
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import plot_confusion_matrix


%matplotlib inline
plt.rcParams['figure.figsize'] = (12,6)
pd.pandas.set_option('display.max_columns',None)

df = pd.read_csv("../input/heart-failure-prediction/heart.csv")
df.head()
```

Building an Machine Learning Model to Predict wether a person as an Heart Disease.
We have 11 Features and 1 Target Variable (Heart Disease)

#### **Getting the Unique Values for Datatype Object**

```python
for i in df.columns:
    if df[i].dtype == "O":
        print(df[i].unique())
```

##### **Changing the Categorical values to Numerical values using get_dummies**

```python
df = pd.get_dummies(df,drop_first=True)
df.head()
```

```python
df.describe()
```

```python
sns.heatmap(df.corr(),annot = True)
plt.show()
```

#### **Train Test & Split**

```python
features = df.drop(columns = "HeartDisease",axis = 1)
label = df["HeartDisease"]
```

```python
features_train,features_test, label_train, label_test = train_test_split(features,label, test_size = 0.1, random_state = 0)
```

```python
clf = GaussianNB()
```

```python
clf.fit(features_train,label_train)
pred = clf.predict(features_test)
print(round(accuracy_score(pred,label_test),2))
```

```python
plot_confusion_matrix(clf, features_test, label_test)
plt.show()
```
